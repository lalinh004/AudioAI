```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

# Khởi tạo trình duyệt Chrome
driver = webdriver.Chrome()
driver.get("https://www.youtube.com/results?search_query=gun+sounds")

# Tạo danh sách để lưu link video
videos = []
scroll_pause_time = 2  # Thời gian chờ để trang tải thêm nội dung

# Tự động cuộn trang để lấy thêm nội dung cho đến khi đủ 100 link
while len(videos) < 100:
    # Cuộn xuống cuối trang
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    
    # Chờ để trang tải thêm nội dung
    time.sleep(scroll_pause_time)
    
    # Lấy lại nội dung của trang và phân tích
    content = driver.page_source
    soup = BeautifulSoup(content, "html.parser")
    
    # Tìm các link video mới
    for video in soup.find_all("a", id="video-title"):
        link = "https://www.youtube.com" + video["href"]
        title = video["title"]
        
        # Thêm video vào danh sách nếu chưa có và danh sách chưa đủ 100 link
        if link not in [v['link'] for v in videos] and len(videos) < 100:
            videos.append({ "link": link, "title": title})
    
    # Nếu đã thu thập đủ 100 link, thoát khỏi vòng lặp
    if len(videos) >= 100:
        break

# Đóng trình duyệt
driver.quit()

# In ra danh sách các link video
# for video in videos:
#     print("Title:", video["title"])
#     print("Link:", video["link"])

# Lưu vào file CSV nếu cần
import csv
with open("youtube_links.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["link", "title"])
    writer.writeheader()
    writer.writerows(videos)

print(f"Đã lưu {len(videos)} link video vào file youtube_links.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
import csv

# Danh sách từ khóa cần tìm kiếm
search_terms = ["Explosion", "blast", "detonation", "blowout", "burst"]

# Tạo danh sách để lưu thông tin video
video_list =[]
scroll_pause_time = 2  # Thời gian chờ để trang tải thêm nội dung

# Khởi tạo trình duyệt Chrome
driver = webdriver.Chrome()

for term in search_terms:
    # Thực hiện tìm kiếm từ khóa
    driver.get(f"https://www.youtube.com/results?search_query={term}")
    videos = []
    # Tự động cuộn trang để lấy thêm nội dung cho đến khi đủ 100 video
    while len(videos) < 100:
        # Cuộn xuống cuối trang
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        
        # Chờ để trang tải thêm nội dung
        time.sleep(scroll_pause_time)
        
        # Lấy lại nội dung của trang và phân tích
        content = driver.page_source
        soup = BeautifulSoup(content, "html.parser")
        
        # Tìm các video mới
        for video in soup.find_all("a", id="video-title"):
            href = video["href"]
            # title = video["title"]
            
            # Lấy id của video từ href
            if 'short' not in href:
                video_id = href.split("v=")[-1]  # Lấy phần sau 'v='
            
            # Thêm video vào danh sách nếu chưa có và danh sách chưa đủ 100 video
            if video_id not in [v['id'] for v in videos] and len(videos) < 100:
                videos.append({"search_term": term, "id": video_id })
        
        # Nếu đã thu thập đủ 100 video, thoát khỏi vòng lặp
        if len(videos) >= 100:
            break

    video_list.extend(videos)

# Đóng trình duyệt
driver.quit()

# In ra danh sách các id video
for video in videos:
    # print("Title:", video["title"])
    print("ID:", video["id"])
    print("Search Term:", video["search_term"])

# Lưu vào file CSV nếu cần
with open("youtube_video_ids.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["search_term", "id" ])
    writer.writeheader()
    writer.writerows(video_list)

print(f"Đã lưu {len(video_list)} ID video vào file youtube_video_ids.csv")
```

```python
import csv
import os
import subprocess

# Hàm tải video từ YouTube sử dụng yt-dlp
def download_video(youtube_id, destination_dir):
    try:
        # Tạo lệnh để tải video từ YouTube với yt-dlp
        command = [
            'yt-dlp',  # Sử dụng yt-dlp thay vì youtube-dl
            '-f', 'best',  # Chọn chất lượng âm thanh tốt nhất
            '-o', f'{destination_dir}/%(id)s.%(ext)s',  # Đường dẫn và tên file: {destination_dir}/{id_video}.{định_dạng}
            f'https://www.youtube.com/watch?v={youtube_id}'  # URL video YouTube
        ]
        
        # Thực thi lệnh tải video
        subprocess.run(command, check=True)
        print(f"Video {youtube_id} đã được tải về thành công.")
    except subprocess.CalledProcessError as e:
        print(f"Không thể tải video {youtube_id}: {e}")

def download_videos_from_csv(csv_file, destination_dir):
    # Kiểm tra nếu thư mục đích chưa tồn tại, tạo mới
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)
        print(f"Tạo thư mục mới: {destination_dir}")
    
    # Đọc file CSV chứa thông tin video
    with open(csv_file, newline='', encoding='utf-8') as file:
        reader = csv.DictReader(file)

        # Lặp qua từng dòng trong CSV và tải video
        for row in reader:
            youtube_id = row['id']  # Lấy ID video từ cột 'id' trong CSV
            download_video(youtube_id, destination_dir)

if __name__ == '__main__':
    # Đường dẫn đến file CSV chứa ID video và thư mục đích để lưu video
    csv_file = 'youtube_video_ids.csv'  # Đảm bảo file CSV của bạn có cột 'id' chứa ID video
    destination_dir = 'downloaded_videos'  # Thư mục để lưu video tải về

    download_videos_from_csv(csv_file, destination_dir)
```

```python
import csv
import os
import subprocess

# Hàm tải video từ YouTube sử dụng yt-dlp
def download_video(youtube_id, destination_dir):
    try:
        # Tạo lệnh để tải video từ YouTube với yt-dlp
        command = [
            'yt-dlp',
            '-f', 'bestaudio',
            '-x',
            '--audio-format', 'wav',
            '--ffmpeg-location', r'C:\ffmpeg\bin\ffmpeg.exe',  # Thay đổi /path/to/ffmpeg thành đường dẫn thực tế
            '-o', f'{destination_dir}/%(id)s.%(ext)s',
            f'https://www.youtube.com/watch?v={youtube_id}'
        ]
        
        # Thực thi lệnh tải video
        result = subprocess.run(command, check=True, capture_output=True, text=True)
        print(f"Video {youtube_id} đã được tải về thành công.")
        
    except subprocess.CalledProcessError as e:
        print(f"Không thể tải video {youtube_id}: {e}")
        print("Chi tiết lỗi:", e.stderr)  # Hiển thị thông tin chi tiết lỗi nếu có

def download_videos_from_csv(csv_file, destination_dir):
    # Kiểm tra nếu thư mục đích chưa tồn tại, tạo mới
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)
        print(f"Tạo thư mục mới: {destination_dir}")
    
    # Đọc file CSV chứa thông tin video
    with open(csv_file, newline='', encoding='utf-8') as file:
        reader = csv.DictReader(file)

        # Lặp qua từng dòng trong CSV và tải video
        for row in reader:
            youtube_id = row['id']  # Lấy ID video từ cột 'id' trong CSV
            download_video(youtube_id, destination_dir)

if __name__ == '__main__':
    # Đường dẫn đến file CSV chứa ID video và thư mục đích để lưu video
    csv_file = 'youtube_video_ids.csv'  # Đảm bảo file CSV của bạn có cột 'id' chứa ID video
    destination_dir = 'downloaded_videos'  # Thư mục để lưu video tải về

    download_videos_from_csv(csv_file, destination_dir)
```
